```python
# 11. Display the information of the employees who have joined in Quarter 1 and 4
for i in dataset.index:
  if (dataset['Quarter of Joining'][i]=='Q1')|(dataset['Quarter of Joining'][i]=='Q4'):
    print(dataset.iloc[i])
```

```python
# 12. How many people are from the southern region joined on Monday?
j=0
for i in dataset.index:
  if (dataset['Short DOW'][i]=='Mon')&(dataset['Region'][i]=='South'):
    j=j+1
print(j)
```

```python
# 13. Display the information of the employees who have more than 13 years experience in the company
for i in dataset.index:
  if(dataset['Age in Company (Years)'][i]>12):
    print(dataset.iloc[i])
```

```python
# 14. There is a lot of redundant data. For example, the information in the below columns is also available in the Date of Joining column.
'''month of Joining
year of Joining
Day of Joining
There could be other such columns containing redundant data. Please identify such columns and take appropriate action.
You can refer to the below link :
https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.drop.html'''
dataset.drop(columns=['Month of Joining'])
dataset.drop(columns=['Year of Joining'])
dataset.drop(columns=['Day of Joining'])
dataset.drop(columns=['Place Name'])
```

```python
# 15. Remove "%" sign from the "Last % Hike" column values of the dataset
dataset['Last % Hike'].str.replace('%','')
```

```python
# 16. How many Female employees got hike more than 20%?
j=0
for i in dataset.index:
  if (dataset['Gender'][i]=='F')&(dataset['Last % Hike'][i]>'20'):
    j=j+1
print(j)
```
